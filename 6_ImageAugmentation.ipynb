In Keras, image augmentation can be done by configuring a number of random transformations to be performed on the images read by the ImageDataGenerator instance.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
    )
```

```python
file = input("Enter the file path: ")
img = image.load_img(file, target_size=(150, 150))
plt.imshow(img)
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
print(x.shape)
```

```python
N = int(input("Enter the number of images to be generated: "))
for i in range(N):
  batch = datagen.flow(x, batch_size=1)
  print(batch[0].shape)
  img = image.array_to_img(batch[0][0]) # Numpy array to PIL image object
  plt.figure(i)
  imgplot = plt.imshow(img)
  plt.axis('off')
plt.show()
```
